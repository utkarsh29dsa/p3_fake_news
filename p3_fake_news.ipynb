```python
import pandas as pd
import numpy as np
```

```python
fake=pd.read_csv('/content/Fake.csv')
```

```python
true=pd.read_csv('/content/True.csv')
```

```python
fake.head()
```

```python
true.head()
```

```python
fake['label']=0
```

```python
true['label']=1
```

```python
fake.head()
```

```python
true.head()
```

```python
news=pd.concat([fake,true], axis=0)
```

```python
news.head()
```

```python
news.tail()
```

```python
print(news.isnull().sum())
```

```python
news=news.drop(['title','subject','date'],axis=1)
```

```python
news.head()
```

```python
news=news.sample(frac=1)
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'],axis=1,inplace=True)
```

```python
news.head()
```

```python
import re
```

```python
def wordopt(text):
  #Convert into lowercase
  text=text.lower()

  #Remove URLS
  text=re.sub(r'https?://\S+|www\.\S+','', text)

  #Remove HTML tags
  text=re.sub(r'<.*?>','', text)

  #Remove punctuation
  text=re.sub(r'[^\w\s]', '', text)

  #Remove newline characters
  text=re.sub(r'\n', ' ', text)

  return text
```

```python
news['text']=news['text'].apply(wordopt)
```

```python
print(news['text'])
```

```python
x=news['text']
y=news['label']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization=TfidfVectorizer()
```

```python
xv_train=vectorization.fit_transform(x_train)
```

```python
xv_test=vectorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
```

```python
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC=DecisionTreeClassifier()
```

```python
DTC.fit(xv_train,y_train)
```

```python
pred_dtc=DTC.predict(xv_test)
```

```python
DTC.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(xv_train,y_train)
```

```python
predict_rfc=rfc.predict(xv_test)
```

```python
rfc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,predict_rfc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc=GradientBoostingClassifier()
```

```python
gbc.fit(xv_train,y_train)
```

```python
pred_gbc=gbc.predict(xv_test)
```

```python
gbc.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gbc))
```

```python
def outputlabel(n):
  if n==0:
    return "Fake News"
  else:
    return "True News"
```

```python
def manualtesting(news):
  testing_news={'text':[news]}
  new_def_test=pd.DataFrame(testing_news)
  new_def_test['text']=new_def_test['text'].apply(wordopt)
  new_x_test=new_def_test['text']
  new_xv_test=vectorization.transform(new_x_test)
  pred_lr=LR.predict(new_xv_test)
  pred_dtc=DTC.predict(new_xv_test)
  pred_rfc=rfc.predict(new_xv_test)
  pred_gbc=gbc.predict(new_xv_test)
  return print("\n\nLR Prediction: {} \nDTC Prediction: {} \nRFC Prediction: {} \nGBC Prediction {} ".format(outputlabel(pred_lr[0]),outputlabel(pred_dtc[0]),outputlabel(pred_rfc[0]),outputlabel(pred_gbc[0])))
```

```python
news_article=str(input())
```

```python
manualtesting(news_article)
```
